#Central Tendency: the location of the middle values
#    1. Mean: Average of a range of values
#    2. Median: Middle value of a range of values
#    3. Mode: Most frequently occurring value

#Spread: how far from the middle the values in a distribution fall
#    1. Variance: Mean of the squares of the individual deviations
#    2. Standard Deviation: Square Root of the variance
#    3. Interquartile Range: Q3 value - Q4 value

#Skewness: Measure of symmetry in the data

#Kurtosis: Measure of how much of the data falls on the tails/edges of the distribution
